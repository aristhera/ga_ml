```r
set.seed(144)
library(caTools)
split = sample.split(parole$violator, SplitRatio = 0.7)
train = subset(parole, split == TRUE)
test = subset(parole, split == FALSE)
```

```r
table(train$violator)
table(test$violator)
```

```r
ViolatorLog = glm(violator ~ ., data=train, family=binomial)
```

```r
summary(ViolatorLog)
```

```r
predictions = predict(ViolatorLog, newdata=test, type="response")
```

```r
summary(predictions)
```

```r
# aufg 5
table(test$violator, predictions>=0.5)
```

```r
tp=14
fn=9
tn=160
fp=19
sensitivity=tp/(tp+fn)
sensitivity
specificity=tn/(tn+fp)
specificity
accuracy=(tn+tp)/(tp+fn+tn+fp)
accuracy
```
