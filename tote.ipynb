```python
tote=10391 # Stand 31.10
x=1000000
tage=1
doppel=11.5 #laut sz.de, 31.10
dici={}
```

```python
while tote < x:
    tote*=2
    tage+=doppel
    dici.update({tage: tote})
```

```python
print(dici)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(list(dici.keys()), list(dici.values()))
plt.ylabel("Tote")
plt.xlabel("Tage")
plt.title("Entwicklung Tote Corona")
plt.show()
```
