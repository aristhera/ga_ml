```python
import csv
import matplotlib.pyplot as plt
from collections import OrderedDict
from operator import itemgetter
filename = '/home/gebhard/Dropbox/s1info/weitz/Notebooks/corona.csv'
with open(filename) as corona:
    reader = csv.reader(corona, delimiter=',')
    header_row = next(reader)
    for index, column_header in enumerate(header_row):
        print(index, column_header)
    staat, faelle = [], []
    for row in reader:
        staat.append(row[1])
        faelle.append(row[2])
        res=dict(zip(staat, faelle))
        result=OrderedDict(sorted(res.items(), key = itemgetter(1), reverse = True))
    print(result)
```

```python
fig = plt.figure(dpi=128, figsize=(10, 6))
result
plt.plot(list(result.keys()), list(result.values()))

plt.title("Corona Wachstumsraten", fontsize=14)
plt.xlabel('Staaten', fontsize=14)

fig.autofmt_xdate()
plt.ylabel("Bestätigte Fälle", fontsize=14)
plt.tick_params(axis='both', which='major')

plt.show()
```
