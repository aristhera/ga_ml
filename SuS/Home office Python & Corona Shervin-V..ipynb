```python
from collections import defaultdict
result=defaultdict(int)
x=1000000
infected=1000
days=1
def zyklen(z):    
    global infected, days
    while infected<x:
        infected*=2
        days+=z
        result[days]+=infected
    return result
print(zyklen(7))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

result
plt.plot(list(result.keys()), list(result.values()))
plt.ylabel("infected")
plt.xlabel("days")
plt.title("Development of corona infections")
plt.show()
```
